```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation

# Simulation: Aerial Camera Coverage Simulation

# Simulation parameters
dt = 1  # Time step (seconds)
total_time = 50  # Total simulation time (seconds)
S = 5  # Speed of the aircraft (units/sec)
FOV_angle = 30  # Field of View (degrees)
gimbal_max_angle = 45  # Max gimbal rotation (degrees)
gimbal_speed = 20  # Gimbal rotation speed (degrees/sec)
camera_forward_tilt = 10  # Camera tilt forward (degrees)
camera_sway_amplitude = 10  # Side-to-side movement amplitude
camera_sway_speed = 1  # Side-to-side movement speed

# Initialize variables
aircraft_x = 0
aircraft_y = 100
aircraft_trail = []
camera_angles = []
covered_areas = []  # Store covered areas

# Create figure
def setup_plot():
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(-100, 500)
    ax.set_ylim(0, 200)
    return fig, ax

fig, ax = setup_plot()
plt.ion()  # Turn on interactive mode

# Function to update the animation
def update(frame):
    global aircraft_x
    ax.clear()
    ax.set_xlim(-100, 500)
    ax.set_ylim(0, 200)
    
    # Move aircraft
    aircraft_x += S * dt
    aircraft_trail.append((aircraft_x, aircraft_y))
    
    # Rotate gimbal (side-to-side movement)
    angle = gimbal_max_angle * np.sin(frame * dt * np.radians(gimbal_speed))
    sway_offset = camera_sway_amplitude * np.sin(frame * dt * np.radians(camera_sway_speed))
    camera_angles.append(angle)
    
    # Store covered area
    fov_radians = np.radians(FOV_angle)
    gimbal_radians = np.radians(angle + camera_forward_tilt)  # Apply forward tilt
    covered_areas.append((aircraft_x + sway_offset, aircraft_y, gimbal_radians))
    
    # Plot aircraft
    ax.scatter(aircraft_x, aircraft_y, c='blue', label="Aircraft")
    
    # Plot trail
    if len(aircraft_trail) > 1:
        trail_x, trail_y = zip(*aircraft_trail)
        ax.plot(trail_x, trail_y, 'b--')
    
    # Plot all covered areas
    for x, y, gimbal_angle in covered_areas:
        arc = patches.Wedge((x, y), 50, np.degrees(gimbal_angle - fov_radians/2),
                             np.degrees(gimbal_angle + fov_radians/2), color='red', alpha=0.3)
        ax.add_patch(arc)
    
    ax.set_title(f"Time: {frame * dt:.1f}s, Gimbal Angle: {angle:.1f}° (Tilted Forward)")
    ax.legend()
    
    plt.draw()
    plt.pause(0.01)

# Run animation in Jupyter Notebook
from IPython.display import display, clear_output

display(fig)
for frame in range(int(total_time / dt)):
    update(frame)
    clear_output(wait=True)
    display(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation

# Simulation: Aerial Camera Coverage Simulation

# Simulation parameters
dt = 1  # Time step (seconds)
total_time = 50  # Total simulation time (seconds)
S = 5  # Speed of the aircraft (units/sec)
FOV_angle = 30  # Field of View (degrees)
gimbal_max_angle = 45  # Max gimbal rotation (degrees)
gimbal_speed = 20  # Gimbal rotation speed (degrees/sec)
camera_forward_tilt = 10  # Camera tilt forward (degrees)
camera_sway_amplitude = 10  # Side-to-side movement amplitude
camera_sway_speed = 1  # Side-to-side movement speed
figure_eight_amplitude = 10  # Amplitude of figure-eight motion
figure_eight_frequency = 2  # Frequency of figure-eight motion

# Initialize variables
aircraft_x = 0
aircraft_y = 100
aircraft_trail = []
camera_angles = []
covered_areas = []  # Store covered areas

# Create figure
def setup_plot():
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(-100, 500)
    ax.set_ylim(0, 200)
    return fig, ax

fig, ax = setup_plot()
plt.ion()  # Turn on interactive mode

# Function to update the animation
def update(frame):
    global aircraft_x
    ax.clear()
    ax.set_xlim(-100, 500)
    ax.set_ylim(0, 200)
    
    # Move aircraft
    aircraft_x += S * dt
    aircraft_trail.append((aircraft_x, aircraft_y))
    
    # Gimbal performs figure-eight motion
    angle = gimbal_max_angle * np.sin(frame * dt * np.radians(gimbal_speed))
    sway_offset = camera_sway_amplitude * np.sin(frame * dt * np.radians(camera_sway_speed))
    figure_eight_offset = figure_eight_amplitude * np.sin(frame * dt * np.radians(figure_eight_frequency)) * np.cos(frame * dt * np.radians(figure_eight_frequency))
    camera_angles.append(angle)
    
    # Store covered area
    fov_radians = np.radians(FOV_angle)
    gimbal_radians = np.radians(angle + camera_forward_tilt)  # Apply forward tilt
    covered_areas.append((aircraft_x + sway_offset + figure_eight_offset, aircraft_y, gimbal_radians))
    
    # Plot aircraft
    ax.scatter(aircraft_x, aircraft_y, c='blue', label="Aircraft")
    
    # Plot trail
    if len(aircraft_trail) > 1:
        trail_x, trail_y = zip(*aircraft_trail)
        ax.plot(trail_x, trail_y, 'b--')
    
    # Plot all covered areas
    for x, y, gimbal_angle in covered_areas:
        arc = patches.Wedge((x, y), 50, np.degrees(gimbal_angle - fov_radians/2),
                             np.degrees(gimbal_angle + fov_radians/2), color='red', alpha=0.3)
        ax.add_patch(arc)
    
    ax.set_title(f"Time: {frame * dt:.1f}s, Gimbal Angle: {angle:.1f}° (Tilted Forward, Figure-8 Motion)")
    ax.legend()
    
    plt.draw()
    plt.pause(0.01)

# Run animation in Jupyter Notebook
from IPython.display import display, clear_output

display(fig)
for frame in range(int(total_time / dt)):
    update(frame)
    clear_output(wait=True)
    display(fig)
```
